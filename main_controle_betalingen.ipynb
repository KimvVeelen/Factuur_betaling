```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import re
from datetime import datetime
from facturen_betalingen import factuurgegevens_opzoeken
from facturen_betalingen import betalingen_opzoeken
from facturen_betalingen import crossreverance
```

```python
facturatie = pd.read_excel("Declaratieoverzicht_dummy.xlsx")
factuurnr = facturatie["Factuur nummer"]
bankbetalingen = pd.read_excel("Bankbijschriften_dummy.xlsx")

facturatie.head()
```

```python
#We maken een dataframe. 
betalingsdataframe = pd.DataFrame(betalingen_opzoeken(bankbetalingen), columns=["Factuurnr", "Bedrag", "Betaaldatum", "Regel in document"])
factuurdataframe = pd.DataFrame(factuurgegevens_opzoeken(facturatie), columns=["Factuurnr", "Bedrag", "Fact datum", "Status", "Klant"])

#betalingsgegevens.to_excel("Betalingsgegevens.xlsx", index=False)
factuurdataframe.head(5)
betalingsdataframe.head(5)
factuurdataframe.to_excel("Factuurdataframe.xlsx", index=False)
```

```python
definitief_overzicht_dataframe = pd.DataFrame(crossreverance(betalingsdataframe, factuurdataframe), columns=["Factuur nr op factuur", "Bedrag op factuur", "Bedrag betaald", "Klant", "Datum factuur", "Datum betaling", "Status", "Regel in betalingsdocument"])

definitief_overzicht_dataframe.head()
```

```python
datum = datetime.now().strftime("%d-%m-%Y_%H_%M")
bestandsnaam = "definitief_Overzicht{date}.xlsx".format(date=datum)
definitief_overzicht_dataframe.to_excel(bestandsnaam, index=False)
print("Excel document is aangemaakt")
```
